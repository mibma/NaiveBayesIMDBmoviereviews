```python
dataset['sentiment'].replace({'positive':1,'negative':0},inplace=True)
```

```python
import re
```

```python
clean=re.compile('<,*?>')
re.sub(clean,'',dataset.iloc[2].review)
```

```python
#function
def clean_html_tags(text):
    clean=re.compile('<,*?>')
    re.sub(clean,'',text)
```

```python
dataset['review'].apply(clean_html_tags)
```

```python
def lower(text):
    return text.lower()
```

```python
dataset['review'].apply(lower)
```

```python
def remove_special(text):
    x=''
    for i in text:
        if i.isalnum():
            x=x+i
        else:
            x=x+' '
    return x       
        
```

```python
dataset['review'].apply(remove_special)
```

```python
#stop words
import nltk
```

```python
from nltk.corpus import stopwords
```

```python
def rem_stopwords(text):
    x=[]
    for i in text.split():
        if i not in stopwords.words('english'):
             x.append(i)
    y=x[:]
    x.clear()
    return y
```

```python
dataset1=dataset.iloc[:5000]
dataset2=dataset.iloc[5000:10000]
dataset3=dataset.iloc[10000:15000]
dataset4=dataset.iloc[25000:30000]
dataset5=dataset.iloc[45000:]

```

```python
dataset1['review'].apply(rem_stopwords)
```

```python
dataset2['review'].apply(rem_stopwords)
```

```python
dataset3['review'].apply(rem_stopwords)
```

```python
dataset4['review'].apply(rem_stopwords)
```

```python
dataset5['review'].apply(rem_stopwords)
```

```python
datasetmer = pd.concat([dataset1, dataset2,dataset3,dataset4,dataset5], ignore_index=True)
```

```python
datasetmer['review']
```

```python
#stemming
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
y=[]
def stem_words(text):
    for i in text:
        y.append(ps.stem(i))
    z=y[:]
    y.clear()
    return z
```

```python
datasetmer['review'].apply(stem_words)
```

```python
def join_back(list_inp):
    return "".join(list_inp)
```

```python
datasetmer['review'].apply(join_back)
```

```python
datasetmer.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=20000)
```

```python
X=cv.fit_transform(datasetmer['review']).toarray()
```

```python
X.shape
```

```python
X[0].max()
```

```python
y=datasetmer.iloc[:,-1].values
```

```python
y
```

```python
y.shape
```

```python
#train test split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.3)
```

```python
X_train.shape
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
clfg=GaussianNB()
clfm=MultinomialNB()
clfb=BernoulliNB()
```

```python
clfg.fit(X_train,y_train)
clfm.fit(X_train,y_train)
clfb.fit(X_train,y_train)
```

```python
y_predg=clfg.predict(X_test)
y_predm=clfm.predict(X_test)
y_predb=clfb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("From Gaussian model",accuracy_score(y_test,y_predg))
print("From Multinomial model",accuracy_score(y_test,y_predm))
print("From Bernoulli model",accuracy_score(y_test,y_predb))

```
